# import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load Dataset

```python
df=pd.read_csv('/content/Titanic-Dataset.csv')
```

# 1. Summary statistics for numeric features

```python
print("Summary statistics for numeric features")
df.describe()
```

# 2. Histograms for numeric features

```python
df.hist(column=['Age','Fare'],bins=20,figsize=(12,5),edgecolor='black')
plt.suptitle('Histograms of Age and Fare')
plt.show()
```

# 3. Boxplots for numeric features grouped by survival

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(x='Survived',y='Age',data=df)
plt.title('Age distribution by Survival')
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,2)
sns.boxplot(x='Survived',y='Fare',data=df)
plt.title('Fare distribution by Survival')
plt.tight_layout()
plt.show()
```

# 4. Pairplot for select numerical and categorical features

```python
sns.pairplot(df,vars=['Age','Fare'],hue='Survived',palette='Set1')
plt.suptitle('Pairplor:Age and Fare by Survival',y=1.02)
plt.show()
```

# 5. Correlation matrix heatmap for numerical columns

```python
corr=df[['Survived','Pclass','Age','SibSp','Parch','Fare']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

# 6. Countplot for categorical features by survival

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.countplot(x='Pclass',hue='Survived',data=df)
plt.title('Survival count by passenger class')
```

#Survival count by sex

```python
plt.subplot(1,3,2)
sns.countplot(x='Sex',hue='Survived',data=df)
plt.title('Survival count by sex')
```

# Survival count by embarkation point

```python
plt.subplot(1,3,3)
sns.countplot(x='Embarked',hue='Survived',data=df)
plt.title('Survival count by embarkation point')
plt.tight_layout()
plt.show()
```

# Basic feature-level inference prints

```python
print("Average Age of Passengers:", df['Age'].mean())
print("Survival Rate:", df['Survived'].mean())

print("Survival rate by Gender:")
print(df.groupby('Sex')['Survived'].mean())

print("Survival rate by Passenger Class:")
print(df.groupby('Pclass')['Survived'].mean())
```
